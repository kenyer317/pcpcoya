## Crear un nueva columna en una posición especifica de un data frame


```python
# Agreguemos una comlumna nueva a la derecha de EN1 de nombre PNE1
# Obtenemos la posición de la columna EN1 le sumamos 1 y esa es la posición requerida
posicion_PE1 = df.columns.get_loc("E1")+1
# Creamos la columna PNE1 = 10% de NE1
df.insert(posicion_PE1, "PE1", value = 10*df["E1"]/100 )
```

## Agregamos las columnas PE2, PE3, PE4 a la dercha de E2, E3, E4, respectivamente


```python
df.insert(df.columns.get_loc("E2")+1, "PE2", value = 20*df["E1"]/100 )
df.insert(df.columns.get_loc("E3")+1, "PE3", value = 30*df["E1"]/100 )
df.insert(df.columns.get_loc("E4")+1, "PE4", value = 40*df["E1"]/100 )
```

```python
df.head()
```

## Agregamos las columnas PP1, PP2, PP3 a la derecha de P1, P2, P3 repectivamente

```python
df.insert(df.columns.get_loc("P1")+1, "PP1", value = 30*df["E1"]/100 )
df.insert(df.columns.get_loc("P2")+1, "PP2", value = 30*df["E1"]/100 )
df.insert(df.columns.get_loc("P3")+1, "PP3", value = 40*df["E1"]/100 )
```

```python
df.head()
```

## Agregamos la columna NT de nota de teoria a la derecha de PE4

```python
df.insert(df.columns.get_loc("PE4")+1, "NT", value = df["PE1"]+ df["PE2"]+ df["PE3"]+ df["PE4"])
```

```python
df.head()
```

## Agregamos la columna NP de nota de práctica a la derecha de PP3

```python
df.insert(df.columns.get_loc("PP3")+1, "NP", value = df["PP1"]+ df["PP2"]+ df["PP3"])
```

```python
df.head()
```

## Agregamos la columna de nota final NT

```python
df.insert(df.columns.get_loc("NP")+1, "NOTA_FINAL", value = df["NT"]*70/100+ df["NP"]*30/100)
```

```python
df.head(10)
```

## Filtramos los aprobados y reprobados
